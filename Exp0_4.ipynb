# Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function y = (x + 3)^2
def function(x):
    return (x + 3)**2
```

```python
# Define the derivative of the function y' = 2 * (x + 3)
def gradient(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent Algorithm
def gradient_descent(starting_x, learning_rate, iterations):
    x = starting_x
    x_values = [x]
    y_values = [function(x)]

    for i in range(iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        x_values.append(x)
        y_values.append(function(x))
        print(f"Iteration {i+1}: x = {x}, y = {function(x)}")

    return x, function(x), x_values, y_values
```

```python
# Parameters
starting_x = 2                # Starting point
learning_rate = 0.1           # Learning rate
iterations = 30               # Number of iterations
```

```python
# Run Gradient Descent
min_x, min_y, x_vals, y_vals = gradient_descent(starting_x, learning_rate, iterations)

print(f"\nLocal minima occurs at x = {min_x}, y = {min_y}")
```

```python
# Plotting the function and the gradient descent path
x_range = np.linspace(-10, 10, 100)
y_range = function(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='y = (x + 3)^2')
plt.scatter(x_vals, y_vals, color='red', label='Gradient Descent Path')
plt.plot(x_vals, y_vals, color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent to Find Local Minima')
plt.legend()
plt.grid(True)
plt.show()
```
